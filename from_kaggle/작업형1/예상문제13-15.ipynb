## 상관관계 구하기
### 주어진 데이터에서 상관관계를 구하고, quality와의 상관관계가 가장 큰 값과, 가장 작은 값을 구한 다음 더하시오!
### 단, quality와 quality 상관관계 제외, 소수점 둘째 자리까지 반올림하여 계산

- 데이터셋 : ../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv

```python
import pandas as pd

df = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
df.head()
```

```python
# 방법1. 피어슨 상관계수 이용
from scipy.stats import pearsonr

arr = []
for c in df.columns:
    if c != 'quality':
        stat, p_val = pearsonr(df[c], df['quality'])
        print(stat)
        arr.append(abs(stat))          ## 상관관계가 가장 큰 값이라고 했으므로 부호 상관없이 크기만 저장
```

```python
round(max(arr)+min(arr), 2)
```

```python
# 방법2. df.corr() 사용
df.corr()
```

```python
# quality간의 상관관계 제거
df.corr()[:-1]
```

```python
df.corr()['quality']
```

```python
round(abs(df.corr()['quality'][:-1]).max() + abs(df.corr()['quality'][:-1]).min(), 2)
```

## 14. 2개 조건에 따른 상위 값
### city와 f4를 기준으로 f5의 평균값을 구한 다음, f5를 기준으로 상위 7개 값을 모두 더해 출력하시오 (소수점 둘째자리까지 출력)
 - 데이터셋 : basic1.csv 

```python
import pandas as pd

df = pd.read_csv("../data/basic1.csv")
df.head()
```

```python
f5_mean = df.groupby(['city', 'f4'])['f5'].mean()          ## df.groupby(['city', 'f4'])[['f5']].mean(): f5 컬럼이 생김
f5_mean
```

```python
print(f5_mean.values)
```

```python
top7_f5 = list(sorted(f5_mean.values, reverse=True))
print(top7_f5)
```

```python
print(round(sum(top7_f5[:7]), 2))
```

```python
# 방법2. 
import pandas as pd

df = pd.read_csv("../data/basic1.csv")
df.head()

#### city와 f4별 f5의 평균 값 (멀티인덱스 출력)
df = df.groupby(['city', 'f4'])[['f5']].mean()
print(df)

# dataframe 전환 후 상위 7개 출력
df = df.reset_index().sort_values('f5', ascending=False).head(7)
print(df)

# f7의 합계 (소수점 둘째자리까지)
round(df['f5'].sum(), 2)
```

## 15. 슬라이싱 & 조건
### 주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음 f1의 결측치를 중앙값으로 채운다. 그리고 f4가 ISFJ와 f5가 20 이상인 f1의 평균값을 출력하시오!
 - 데이터셋 : basic1.csv 

```python
import pandas as pd

df = pd.read_csv("../data/basic1.csv")
df.head()
```

```python
df = df.sort_values('age', ascending=False)[:20]
df
```

```python
# f1 결측치 중앙값으로 대체
df['f1'] = df['f1'].fillna(df['f1'].median())

# f4가 ISFJ와 f5가 20 이상인 f1의 평균값
f1_mean = df['f1'][(df['f4']=='ISFJ') & (df['f5'] >= 20)].mean()
print(f1_mean)
```
