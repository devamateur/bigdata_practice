## 5. 조건에 맞는 데이터 표준편차 구하기
### 주어진 데이터 중 basic1.csv에서 'f4'컬럼 값이 'ENFJ'와 'INFP'인 'f1'의 표준편차 차이를 절대값으로 구하시오
- 데이터셋 : basic1.csv


```python
import pandas as pd

df = pd.read_csv('../data/basic1.csv')

df.head()
```

```python
# ENFJ, INFP
enfj = df['f1'][(df['f4'] == 'ENFJ')]
enfj
```

```python
infp = df['f1'][(df['f4'] == 'INFP')]
infp
```

```python
# 둘의 표준편차 차이 절대값

abs(enfj.std() - infp.std())
```

## 6. 결측치 제거 및 그룹 합계에서 조건에 맞는 값 찾아 출력
### 주어진 데이터 중 basic1.csv에서 'f1'컬럼 결측 데이터를 제거하고, 'city'와 'f2'을 기준으로 묶어 합계를 구하고, 'city가 경기이면서 f2가 0'인 조건에 만족하는 f1 값을 구하시오
- 데이터셋 : basic1.csv

```python
import pandas as pd

df = pd.read_csv('../data/basic1.csv')

print(df.shape)
df.head()
```

```python
# f1 컬럼 결측 데이터 제거
df = df[~df['f1'].isna()]
```

```python
print(df.shape)
df.head()
```

```python
# city와 f2를 기준으로 묶어서 합계
city_f2_sum = df.groupby(['city', 'f2']).sum()
city_f2_sum
```

```python
# city가 경기이면서 f2가 0'인 조건에 만족하는 f1 값
found = city_f2_sum.iloc[0]['f1']
```

```python
found
```

## 7. 값 변경 및 2개 이상의 조건
### 'f4'컬럼의 값이 'ESFJ'인 데이터를 'ISFJ'로 대체하고, 'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age'컬럼의 최대값을 출력하시오!
- 데이터셋 : basic1.csv

```python
import pandas as pd

df = pd.read_csv('../data/basic1.csv')

df.head(20)
```

```python
# esfj -> isfj

df['f4'] = df['f4'].replace('ESFJ', 'ISFJ')
```

```python
df.iloc[8]
```

```python
# 'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age'컬럼의 최대값
result = df[(df['city'] == '경기') & (df['f4'] == 'ISFJ')]
result
```

```python
max(result['age'])
```

## 8. 누적합 그리고 보간(결측치 처리)
### 주어진 데이터 셋에서 'f2' 컬럼이 1인 조건에 해당하는 데이터의 'f1'컬럼 누적합을 계산한다. 이때 발생하는 누적합 결측치는 바로 뒤의 값을 채우고, 누적합의 평균값을 출력한다. (단, 결측치 바로 뒤의 값이 없으면 다음에 나오는 값을 채워넣는다)
- 데이터셋 : basic1.csv

```python
import pandas as pd

df = pd.read_csv('../data/basic1.csv')

df.head()
```

```python
# f2 컬럼이 1일 때 누적합 계산
# 결측치: 바로 뒤의 값으로 채움, 바로 뒤 값이 없으면 다음에 나오는 값으로 채움
df = df[df['f2'] == 1]['f1'].cumsum()# 누적합
```

```python
df = df.fillna(method='bfill')       # bfil: 바로 뒤의 값으로 채우는 방법
# df.fllna(method='pad'): 이전값으로 채우는 방법
```
